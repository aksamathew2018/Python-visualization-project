This project analyses the employee dataset of ABC Company to understand its workforce structure. The study includes the distribution of employees across teams and positions, identification of the predominant age group, and evaluation of salary expenditure by both team and job role. It also examines the correlation between age and salary. Visualisations such as bar charts, pie charts, histograms, and scatter plots are used to present the findings clearly and effectively.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/ABC Company.xlsx - Sheet1.csv')
df.head(2)
```

```python
df.columns
```

```python
import random
```

```python
#
#Correct the data in the "height" column by replacing it with random
#numbers between 150 and 180.
df['Height']=[random.randint(150,180) for i in range(len(df))]
df.head(2)
```

```python
#1.Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.
df['Team'].value_counts()
```

```python
df['Team'].value_counts().mean()
```

```python
#and calculate the percentage split relative to the total number of employees.
team_counts = df['Team'].value_counts()
team_percentage = (team_counts / team_counts.sum()) * 100
df3 = pd.concat([team_counts, team_percentage], axis=1)
df3.columns = ['No of Employees', 'Percentage']
df3
```


The New Orleans Pelicans have the highest number of employees, while the Minnesota Timberwolves and Orlando Magic have the fewest. Most teams have around 15 members, making the average team size approximately 15.


```python
sns.barplot(x=df3.index,y=df3['No of Employees'],color='red')
plt.title('Distribution of Employees by Team')
plt.xlabel('Team')
plt.ylabel('No of Employees')
plt.show()
```

```python
plt.pie(df3['No of Employees'], labels=df3.index, autopct='%1.1f%%')
plt.title('Distribution of Employees by Team')
plt.show()
```

```python
#Segregate employees based on their positions within the company.
df10=pd.DataFrame(df['Position'].value_counts())
df10.columns=['No of Employees']
df10
```

```python
#Identify the predominant age group among employees.
df6=df.sort_values(by='Age',ascending=True)
df6.head(10)
```

```python
df.nlargest(5,'Age')
```

```python
df.nsmallest(5,'Age')
```

```python
#Identify the predominant age group among employees.
age_range = [19, 26, 32, 41]
labels = ['1925', '2631', '3240']

df['Age_Group'] = pd.cut(df['Age'], bins=age_range, labels=labels, right=False)
df['Age_Group'].value_counts()
```

```python
plt.pie(df['Age_Group'].value_counts(), labels=df['Age_Group'].value_counts().index)
plt.title('Predominant Age Group')
plt.legend()
plt.show()
```

Most employees fall within the 1925 age group. The youngest employee is 19 years old, while the oldest is 40.

```python
#Discover which team and position have the highest salary expenditure.
df8=df[['Position','Salary']]
a=df8.groupby('Position').sum()
a
```

```python
sns.barplot(x=a.index,y=a['Salary'],color='green')
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
plt.pie(a['Salary'], labels=a.index,autopct='%1.1f%%')
plt.title('Salary Distribution across positions')
plt.legend()
plt.show()
```

```python
#Investigate if there's any correlation between age and salary, and represent it visually.
df9=df[['Age','Salary']]
df9.corr()
```

```python
sns.scatterplot(x=df9['Age'],y=df9['Salary'],color='red')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
sns.heatmap(df9.corr(),annot=True)
```

The correlation of 0.21 indicates a weak positive relationship between Age and Salary. This means that as age increases, salary tends to increase slightly.
